```python
import pandas as pd
import numpy as np
```

```python
df_exif = pd.read_pickle('./df_exif.pickle')
df_mysql = pd.read_pickle('./df_mysql.pickle')
df_mysql.keys()
```

```python
df_exif.rename(columns={'FileName':'filename'},inplace=True)
df_exif.keys()
```

```python
df = pd.merge(df_mysql, df_exif, how='inner', on=['filename'])
df.head()
```

```python
df = df.loc[df['hasExifFields']>-1]
len(df)
df.keys()
```

```python
df.groupby(['brand','device','manufacturer','model','Make','Model']).size().sort_values(ascending=False).reset_index().rename(columns={0:'count'})
#tmp.to_csv('model_overview.csv')
```

```python
trimmer = lambda x: x.strip()
df['model'] = df['model'].apply(trimmer)
df['Model'] = df['Model'].apply(trimmer)
df[df['model'] != df['Model']][['Model','model','filename']].groupby(['Model','model']).size().sort_values(ascending=False).reset_index().rename(columns={0:'count'})
```

```python
print('{}/{} mismatches'.format(len(df[df['model'] != df['Model']]),len(df)))
```

```python
def check_key_identity(df,prim_key, sec_key):
    sub = df[df[prim_key] != df[sec_key]]
    for k_1, v_1 in enumerate(sorted(sub[prim_key].unique())):
        print(df[df[prim_key] == v_1].groupby([prim_key,sec_key]).size().sort_values(ascending=False).reset_index().rename(columns={0:'count'}))
        print('\n\n#########')
    
check_key_identity(df,'Model', 'model')
```
